```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from scipy.fft import fft

font = {'size'   : 15}
matplotlib.rc('font', **font)
```

```python
x = np.genfromtxt("Signal.txt", dtype = float)
zarib = 1 / 4000
nyquist = 2000
```

# سوال ۲)
## بخش اول)
### الف)
#### 1)
Sample Rate = 1000
#### 2)
ضریب اول: 1/2

ضریب دوم: 1/2000
## بخش دوم)
### الف)
#### 1)

```python
def fourier(w):
    t = np.linspace(0, 1, 2000)
    tmp = [np.exp(-1j * w * i) for i in t] 
    return np.dot(x, tmp) * zarib

w = np.arange(0, nyquist, 1)
fourier_seri = fourier(w)
```

##### 2)
Nyquist Frequence = 2000

```python
fig = plt.figure()
ax = fig.add_axes([0, .3, 1, 1])
ax.plot(w, np.abs(fourier_seri))
ax.set_xlabel("w", fontsize = 15)
ax.set_ylabel("|X(w)|", fontsize = 15)
fig.supxlabel("Fig. 2,2,1,2: Amplitudes of Fourier Transform", fontsize = 18)
plt.savefig('B_Figs/Fig. 2,2,1,2.png', dpi=300, bbox_inches='tight')
```

### ب)
#### 1)

```python
fig = plt.figure()
ax = fig.add_axes([0, .3, 1, 1])
ax.plot(np.abs(fft(x)) * zarib)
ax.set_xlabel("w", fontsize = 15)
ax.set_ylabel("|X(w)|", fontsize = 15)
fig.supxlabel("Fig. 2,2,2,1: Amplitudes of Fourier Transform using scipy.fft", fontsize = 18)
plt.savefig('B_Figs/Fig. 2,2,2,1.png', dpi=300, bbox_inches='tight')
```

#### 2)
مشکل به خاطر این ایجاد می‌شود که هنگام تبدیل انتگرال به سیگما از dx
صرف نظر می‌شود.
همچنین یک ضریب ۱/۲ هم به خاطر فرکانس های مثبت و منفی لازم است
با اعمال ضریب هایی که در بخش اول گفته شده است مشکل حل می‌شود.

```python
t = np.linspace(0, 1, 2000)
g = np.sin(2 * np.pi * 5 * t)
g = g * 2
g = g + np.sin(np.pi / 2)
G = fft(g)

fig = plt.figure()
ax = fig.add_axes([0, .3, 1, 1])
ax.plot(np.abs(G) * zarib)
ax.set_xlabel("w", fontsize = 15)
ax.set_ylabel("|G(w)|", fontsize = 15)
fig.supxlabel("Fig. 2,2,2,2: Amplitudes of Fourier Transform using scipy.fft", fontsize = 18)
plt.savefig('B_Figs/Fig. 2,2,2,2.png', dpi=300, bbox_inches='tight')
```
