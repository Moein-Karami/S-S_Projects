```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
# font = {'size'   : 22}
# matplotlib.rc('font', **font)
```

# سوال ۱
## بخش اول
### الف)


```python
t = np.linspace(0, 2, 1000);
x = 2 * np.sin(3 * t + np.pi / 3)

fig = plt.figure()
ax = fig.add_axes([0, .3, 1, 1])
ax.plot(t, x)
ax.grid(True, which='both')
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
ax.set_xlabel("t", fontsize = 22)
ax.set_ylabel("x", fontsize = 22)
fig.supxlabel("Fig. 1,1,1: x = 2 * sin (3t + pi/3)", fontsize = 22)
plt.savefig('A_Figs/Fig. 1,1,1.png', dpi=300, bbox_inches='tight')
```

### ب)

```python
x = np.exp(1j * 2 * np.pi * 3 * t)

fig = plt.figure()
ax = fig.add_axes([0, .3, 1, 1])

ax.plot(t, np.real(x), label = 'real part')
ax.plot(t, np.imag(x), label = 'imaginary part')
ax.grid(True, which='both')
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')

ax.set_xlabel("t", fontsize = 22)
ax.set_ylabel("x", fontsize = 22)

fig.legend(bbox_to_anchor = (0.1, 0.25))

fig.supxlabel("Fig. 1,1,2: x = exp(2i * pi * f * t)", fontsize = 22)
plt.savefig('A_Figs/Fig. 1,1,2.png', dpi=300, bbox_inches='tight')
```

### ج)

```python
fig = plt.figure()
ax = fig.gca(projection = '3d')

ax.plot(t, np.real(x), np.imag(x), zdir = 'z')
ax.set_xlabel('t', fontsize = 12)
ax.set_ylabel('real(x)', fontsize = 12)
ax.set_zlabel('imaginary(x)', fontsize = 12)

fig.supxlabel("Fig. 1,1,3: 3D plot", fontsize = 22)
plt.savefig('A_Figs/Fig. 1,1,3.png', dpi=300, bbox_inches='tight')
```

### ج)
#### 1)

```python
t = np.linspace(-1, 1, 2000);
theta = 0
x = np.sin(2 * t * 5 * np.pi + theta) * np.exp(-t**2)

fig = plt.figure()
ax = fig.add_axes([0, .3, 1, 1])
ax.plot(t, x)
ax.grid(True, which='both')
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
ax.set_xlabel("t", fontsize = 22)
ax.set_ylabel("x", fontsize = 22)
fig.supxlabel("Fig. 1,2,3,1: x = sin(2 * pi * 5 * ti + theta) * exp(-t^2)", fontsize = 22)
plt.savefig('A_Figs/Fig. 1,2,3,1.png', dpi=300, bbox_inches='tight')
```

#### 2)
برای فرکانس ۴.۵ این مقدار بیشینه است.

```python
freq = np.arange(2, 10, 0.5)
y = [np.dot(np.sin(i * np.linspace(-1, 1, 2000)), x) for i in freq]

fig = plt.figure()
ax = fig.add_axes([0, .3, 1, 1])
ax.stem(freq, y, use_line_collection = True)
ax.set_xlabel('frequence')

fig.supxlabel("Fig. 1,2,3,2: Dot Product of f(t)", fontsize = 22)
plt.savefig('A_Figs/Fig. 1,2,3,2.png', dpi=300, bbox_inches='tight')
```

#### 3)
باعث تغییر مقادیر در نمودار می‌شود (هم کاهش و هم افزایش)
مشکلی که ممکن است پیش بیاید این است که ممکن است با وارد کردن یک سیگنال در دو زمان متفاوت به علت اختلاف زمان و در نتیجه اختلاف فاز ضرب داخلی متفاوتی به دست بیاوریم در صورتی که ما به ازای دو سیگنال یکسان انتظار دو پاسخ یکسان داریم.

```python
t = np.linspace(-1, 1, 2000);
freq = np.arange(2, 10, 0.5)
theta = np.pi/2
x1 = np.sin(2 * t * 5 * np.pi + theta) * np.exp(-t**2)
y1 = [np.dot(np.sin(i * np.linspace(-1, 1, 2000)), x1) for i in freq]

theta = np.pi/4
x2 = np.sin(2 * t * 5 * np.pi + theta) * np.exp(-t**2)
y2 = [np.dot(np.sin(i * np.linspace(-1, 1, 2000)), x2) for i in freq]


fig = plt.figure(figsize = (20, 6))
ax1 = fig.add_axes([0, .2, .45, 1])
ax2 = fig.add_axes([0.55, .2, .5, 1])

ax1.stem(freq, y1, use_line_collection = True)
ax2.stem(freq, y2, use_line_collection = True)

ax1.set_xlabel('frequency', fontsize = 15)
ax2.set_xlabel('frequency', fontsize = 15)

ax1.set_title('theta = pi / 2', fontsize = 20)
ax2.set_title('theta = pi / 4', fontsize = 20)

fig.supxlabel("Fig. 1,2,3,3: Plot for different values of theta", fontsize = 26)
plt.savefig('A_Figs/Fig. 1,2,3,3.png', dpi=300, bbox_inches='tight')
```

### د)

```python
t = np.linspace(-1, 1, 2000);
freq = np.arange(2, 10, 0.5)
theta = np.pi/2
x1 = np.exp(2 * t * 5 * np.pi + theta) * np.exp(-t**2)
y1 = [np.dot(np.exp(2j * np.pi * i * np.linspace(-1, 1, 2000)), x1) for i in freq]
y1 = np.abs(y1)

theta = np.pi/4
x2 = np.exp(2 * t * 5 * np.pi + theta) * np.exp(-t**2)
y2 = [np.dot(np.exp(2j * np.pi * i * np.linspace(-1, 1, 2000)), x2) for i in freq]
y2 = np.abs(y2)

fig = plt.figure(figsize = (20, 6))
ax1 = fig.add_axes([0, .2, .45, 1])
ax2 = fig.add_axes([0.55, .2, .5, 1])

ax1.stem(freq, y1, use_line_collection = True)
ax2.stem(freq, y2, use_line_collection = True)

ax1.set_xlabel('frequency', fontsize = 15)
ax2.set_xlabel('frequency', fontsize = 15)

ax1.set_title('theta = pi / 2', fontsize = 22)
ax2.set_title('theta = pi / 4', fontsize = 22)

fig.supxlabel("Fig. 1,2,4,1: use exp(2j pi f t) instead of sin", fontsize = 32)
plt.savefig('A_Figs/Fig. 1,2,4,.png', dpi=300, bbox_inches='tight')
```
