## Sequences

```python
import numpy as np

X1 = exp(-j * np.pi * n / 2) * u(n)
X2 = n * u(n)

x1 = X1.seq((0, 5))
x2 = X2.seq((0, 5))

x3 = x1 + x1
x4 = np.zeros(6)
x4 = x4 * 1j;

for m in range(6):
    for i in range(6):
        x4[m] += x2.as_array()[i] * x3.as_array()[5 - m + i]
x4
```

## Discrete Transforms

```python
a = symbol('a')
x = (0.5 - a)**n * u(n) + 10 * (a + 1/3)**n * u(n)
```
