```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

font = {'size'   : 22}
matplotlib.rc('font', **font)
```

# الف)
در این قسمت پاسخ سیستم را به صورت تحلیلی حساب می‌کنیم

```python
def ans(x, alpha = 0.5):
    if x < 0:
        return 0;
    elif x <= 4:
        return np.sum([alpha ** i for i in range(x + 1)])
    else:
        return np.sum([alpha ** i for i in range(x - 4, x + 1)])
```

# ب)

```python
part1 = [ans(i) for i in range(-10, 0)]
part1_lim = np.arange(-10, 0)

part2 = [ans(i) for i in range(0, 5)]
part2_lim = np.arange(0, 5)

part3 = [ans(i) for i in range(5, 16)]
part3_lim = np.arange(5, 16)

fig = plt.figure(figsize = (20, 6))
axes1 = fig.add_axes([0, .24, 0.25, 1]);
axes2 = fig.add_axes([0.33, .24, 0.25, 1]);
axes3 = fig.add_axes([0.66, .24, 0.25, 1]);

axes1.stem(part1_lim, part1, use_line_collection = True)
axes2.stem(part2_lim, part2, use_line_collection = True)
axes3.stem(part3_lim, part3, use_line_collection = True)

fig.supxlabel("Fig. 3,2,1: System's output for different parts", fontsize = 32)
axes1.set_xlabel('n', fontsize = 32)
axes2.set_xlabel('n', fontsize = 32)
axes3.set_xlabel('n', fontsize = 32)
axes1.set_ylabel('y[n]', fontsize = 32)
axes2.set_ylabel('y[n]', fontsize = 32)
axes3.set_ylabel('y[n]', fontsize = 32)
axes1.set_title('n < 0', fontsize = 32)
axes2.set_title('0 <= n <= 4', fontsize = 32)
axes3.set_title('4 < n', fontsize = 32)
plt.savefig("C/Fig. 3,2,1: System's output for different parts.png", dpi=300, bbox_inches='tight')
```

```python
n = np.arange(-5, 16, 1)
y = [ans(i) for i in n]
fig = plt.figure(figsize = (20, 6))
axes1 = fig.add_axes([0, .24, 1, 1]);

axes1.stem(n, y, use_line_collection = True)

fig.supxlabel("Fig. 3,2,2: System's output for complete signal", fontsize = 32)
axes1.set_xlabel('n', fontsize = 32)
axes1.set_ylabel('y[n]', fontsize = 32)
axes1.set_title('y[n] = x[n] * h[n]', fontsize = 32)
plt.savefig('C/Fig. 3,2,2: System\'s output for complete signal.png', dpi=300, bbox_inches='tight')
```
