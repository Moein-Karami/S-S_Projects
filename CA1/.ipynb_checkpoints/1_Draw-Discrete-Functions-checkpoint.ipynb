```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

$A = 10 , K = 2 \implies T = 10$

```python
leng1 = 20
leng2 = 100
n1 = np.arange(0, leng1)
n2 = np.arange(0, leng2)
A, K = 2, 0.1
T = 10
```

```python
font = {'size'   : 22}
matplotlib.rc('font', **font)
```

```python
def x(n):
    return A * np.cos(n * np.pi / 5)
```

# الف)

## X[n]

```python
fig = plt.figure(figsize = (20, 6))
axes1 = fig.add_axes([0, .2, 0.45, 1]);
axes2 = fig.add_axes([0.55, .20, 0.5, 1]);
axes1.stem(n1, x(n1), use_line_collection = True)
axes2.stem(n2, x(n2), use_line_collection = True)

fig.supxlabel("Fig. 1,0: Plot of x[n]", fontsize = 32)
axes1.set_xlabel('n = 20', fontsize = 32)
axes2.set_xlabel('n = 100', fontsize = 32)
axes1.set_ylabel('x[n]', fontsize = 32)
axes2.set_ylabel('x[n]', fontsize = 32)
plt.savefig('A/Fig. 1,0: Plot of x[n].png', dpi=300, bbox_inches='tight')
```

## A

```python
y1 = 1/T * (x(n1 + 1) - x(n1))
y2 = 1/T * (x(n2 + 1) - x(n2))

fig = plt.figure(figsize = (20, 6))
axes1 = fig.add_axes([0, .2, 0.45, 1]);
axes2 = fig.add_axes([0.55, .20, 0.5, 1]);
axes1.stem(n1, y1, use_line_collection = True)
axes2.stem(n2, y2, use_line_collection = True)

axes1.set_xlabel('n = 20', fontsize = 32)
axes2.set_xlabel('n = 100', fontsize = 32)
axes1.set_ylabel('y[n]', fontsize = 32)
axes2.set_ylabel('y[n]', fontsize = 32)
fig.supxlabel("Fig. 1,1: Plot of y[n] for part A", fontsize = 32)
plt.savefig('A/Fig. 1,1: Plot of y[n] for part A.png', dpi=300, bbox_inches='tight')
```

## B

```python
y1 = 1 / (2 * T) * (x(n1 + 1) - x(n1 - 1))
y2 = 1 / (2 * T) * (x(n2 + 1) - x(n2 - 1))

fig = plt.figure(figsize = (20, 6))
axes1 = fig.add_axes([0, .2, 0.45, 1]);
axes2 = fig.add_axes([0.55, .20, 0.5, 1]);
axes1.stem(n1, y1, use_line_collection = True)
axes2.stem(n2, y2, use_line_collection = True)

axes1.set_xlabel('n = 20', fontsize = 32)
axes2.set_xlabel('n = 100', fontsize = 32)
axes1.set_ylabel('y[n]', fontsize = 32)
axes2.set_ylabel('y[n]', fontsize = 32)
fig.supxlabel("Fig. 1,2: Plot of y[n] for part B", fontsize = 32)
plt.savefig('A/Fig. 1,2: Plot of y[n] for part B.png', dpi=300, bbox_inches='tight')
```

## C

```python
y1 = 1 / (T * T) * (x(n1 + 1) - 2 * x(n1) + x(n1 - 1))
y2 = 1 / (T * T) * (x(n2 + 1) - 2 * x(n2) + x(n2 - 1))

fig = plt.figure(figsize = (20, 6))
axes1 = fig.add_axes([0, .2, 0.45, 1]);
axes2 = fig.add_axes([0.55, .20, 0.5, 1]);
axes1.stem(n1, y1, use_line_collection = True)
axes2.stem(n2, y2, use_line_collection = True)

axes1.set_xlabel('n = 20', fontsize = 32)
axes2.set_xlabel('n = 100', fontsize = 32)
axes1.set_ylabel('y[n]', fontsize = 32)
axes2.set_ylabel('y[n]', fontsize = 32)
fig.supxlabel("Fig. 1,3: Plot of y[n] for part C", fontsize = 32)
plt.savefig('A/Fig. 1,3: Plot of y[n] for part C.png', dpi=300, bbox_inches='tight')
```

## D

```python
y1 = 0.1 * x(n1 - 2) + 0.2 * x(n1 - 1) + 0.4 * x(n1) + 0.2 * x(n1 + 1) + 0.1 * x(n1 + 2)
y2 = 0.1 * x(n2 - 2) + 0.2 * x(n2 - 1) + 0.4 * x(n2) + 0.2 * x(n2 + 1) + 0.1 * x(n2 + 2)

fig = plt.figure(figsize = (20, 6))
axes1 = fig.add_axes([0, .2, 0.45, 1]);
axes2 = fig.add_axes([0.55, .20, 0.5, 1]);
axes1.stem(n1, y1, use_line_collection = True)
axes2.stem(n2, y2, use_line_collection = True)

axes1.set_xlabel('n = 20', fontsize = 32)
axes2.set_xlabel('n = 100', fontsize = 32)
axes1.set_ylabel('y[n]', fontsize = 32)
axes2.set_ylabel('y[n]', fontsize = 32)
fig.supxlabel("Fig. 1,4: Plot of y[n] for part D", fontsize = 32)
plt.savefig('A/Fig. 1,4: Plot of y[n] for part D.png', dpi=300, bbox_inches='tight')
```
