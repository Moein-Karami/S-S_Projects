```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

font = {'size'   : 22}
matplotlib.rc('font', **font)
```

```python
def delta(n):
    return np.piecewise(n, [n == 0, n != 0], [1, 0])
```

```python
n = np.arange(-10, 100, 1)
```

## numpy.cumsum()
برای تبدیل پاسخ ضربه به پاسخ تابع پله از این تابع استفاده شده که به ازای آرایه ی ورودی یک آرایه که عنصر 
$i$
ام آن معادل با جمع 
$i$
عنصر اول است،‌ می‌گراند

# الف)
پاسخ تحلیلی داخل فایل
$Hand Writings$
موجود است

## System(A)

```python
y1 = 0.5 * (delta(n) + delta(n - 1))
y2 = np.cumsum(y1);

fig = plt.figure(figsize = (20, 6))
axes1 = fig.add_axes([0, .24, 0.45, 1]);
axes2 = fig.add_axes([0.55, .24, 0.5, 1]);
axes1.stem(n, y1, use_line_collection = True)
axes2.stem(n, y2, use_line_collection = True)

fig.supxlabel("Fig. 2,1,1: System(A) output for pulse and step functions", fontsize = 32)
axes1.set_xlabel('n', fontsize = 32)
axes2.set_xlabel('n', fontsize = 32)
axes1.set_ylabel('y[n]', fontsize = 32)
axes2.set_ylabel('y[n]', fontsize = 32)
axes1.set_title('x[n] = delta(n)', fontsize = 32)
axes2.set_title('x[n] = u[n]', fontsize = 32)
plt.savefig('B_Figs/Fig. 2,1,1: System(A) output for pulse and step functions.png', dpi=300, bbox_inches='tight')
```

## System(B)

```python
y1 = (delta(n + 1) - delta(n))
y2 = np.cumsum(y1);

fig = plt.figure(figsize = (20, 6))
axes1 = fig.add_axes([0, .24, 0.45, 1]);
axes2 = fig.add_axes([0.55, .24, 0.5, 1]);
axes1.stem(n, y1, use_line_collection = True)
axes2.stem(n, y2, use_line_collection = True)

fig.supxlabel("Fig. 2,1,2: System(B) output for pulse and step functions", fontsize = 32)
axes1.set_xlabel('n', fontsize = 32)
axes2.set_xlabel('n', fontsize = 32)
axes1.set_ylabel('y[n]', fontsize = 32)
axes2.set_ylabel('y[n]', fontsize = 32)
axes1.set_title('x[n] = delta(n)', fontsize = 32)
axes2.set_title('x[n] = u[n]', fontsize = 32)
plt.savefig('B_Figs/Fig. 2,1,2: System(B) output for pulse and step functions.png', dpi=300, bbox_inches='tight')
```

```python
y1 = (delta(n + 1) - 2 * delta(n) + delta(n - 1))
y2 = np.cumsum(y1);
fig = plt.figure(figsize = (20, 6))
axes1 = fig.add_axes([0, .24, 0.45, 1]);
axes2 = fig.add_axes([0.55, .24, 0.5, 1]);
axes1.stem(n, y1, use_line_collection = True)
axes2.stem(n, y2, use_line_collection = True)

fig.supxlabel("Fig. 2,1,3: System(C) output for pulse and step functions", fontsize = 32)
axes1.set_xlabel('n', fontsize = 32)
axes2.set_xlabel('n', fontsize = 32)
axes1.set_ylabel('y[n]', fontsize = 32)
axes2.set_ylabel('y[n]', fontsize = 32)
axes1.set_title('x[n] = delta(n)', fontsize = 32)
axes2.set_title('x[n] = u[n]', fontsize = 32)
plt.savefig('B_Figs/Fig. 2,1,3: System(C) output for pulse and step functions.png', dpi=300, bbox_inches='tight')
```

## System(D)

```python
y1 = [np.cos(np.pi * i / 10) * (0 <= i <= 19) for i in n]
y2 = np.cumsum(y1)

fig = plt.figure(figsize = (20, 6))
axes1 = fig.add_axes([0, .24, 0.45, 1]);
axes2 = fig.add_axes([0.55, .24, 0.5, 1]);
axes1.stem(n, y1, use_line_collection = True)
axes2.stem(n, y2, use_line_collection = True)

fig.supxlabel("Fig. 2,1,4: System(C) output for pulse and step functions", fontsize = 32)
axes1.set_xlabel('n', fontsize = 32)
axes2.set_xlabel('n', fontsize = 32)
axes1.set_ylabel('y[n]', fontsize = 32)
axes2.set_ylabel('y[n]', fontsize = 32)
axes1.set_title('x[n] = delta(n)', fontsize = 32)
axes2.set_title('x[n] = u[n]', fontsize = 32)
plt.savefig('B_Figs/Fig. 2,1,4: System(C) output for pulse and step functions.png', dpi=300, bbox_inches='tight')
```

# ب)
توجه کنید که در بالا پاسخ تابع پله به ازای
$m_u = \inf$
به کمک تابع
$numpy.cumsum()$
کشیده شده است و همین توابع پیش‌بینی ما در حالتی می‌باشند که 
$m_d$
به
$inf$
میل می‌کند، حال این توابع را به ازای
$m_u$
های کوچک تر رسم می‌کنیم.

## System(A)

```python
y1 = 0.5 * (delta(n) + delta(n - 1))
y2 = [np.sum(y1[max(0, i - 5) : i]) for i in range(0, len(y1))]
y3 = [np.sum(y1[max(0, i - 15) : i]) for i in range(0, len(y1))]
y4 = [np.sum(y1[max(0, i - 20) : i]) for i in range(0, len(y1))]

fig = plt.figure(figsize = (20, 6))
axes1 = fig.add_axes([0, .24, 0.25, 1]);
axes2 = fig.add_axes([0.33, .24, 0.25, 1]);
axes3 = fig.add_axes([0.66, .24, 0.25, 1]);

axes1.stem(n, y2, use_line_collection = True)
axes2.stem(n, y3, use_line_collection = True)
axes3.stem(n, y4, use_line_collection = True)

fig.supxlabel("Fig. 2,2,1: System(A) output for pulse and step functions", fontsize = 32)
axes1.set_xlabel('n', fontsize = 32)
axes2.set_xlabel('n', fontsize = 32)
axes3.set_xlabel('n', fontsize = 32)
axes1.set_ylabel('y[n]', fontsize = 32)
axes2.set_ylabel('y[n]', fontsize = 32)
axes3.set_ylabel('y[n]', fontsize = 32)
axes1.set_title('m_d = 5', fontsize = 32)
axes2.set_title('m_d = 15', fontsize = 32)
axes3.set_title('m_d = 20', fontsize = 32)
plt.savefig('B_Figs/Fig. 2,2,1: System(A) output for pulse and step functions.png', dpi=300, bbox_inches='tight')
```
